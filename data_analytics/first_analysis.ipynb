```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Փոքր dataset (սինթետիկ տվյալներ)
data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    "Income": [200000, 300000, 450000, 600000, 800000],
    "Debt": [1000000, 850000, 700000, 500000, 300000]
}

df = pd.DataFrame(data)

print("Տվյալներ")
print(df)

# Գրաֆիկ
plt.plot(df["Year"], df["Income"], marker="o", label="Income")
plt.plot(df["Year"], df["Debt"], marker="s", label="Debt", linestyle="--")
plt.title("Ապագայի ճանապարհ — Եկամուտ vs Պարտք")
plt.xlabel("Տարի")
plt.ylabel("Գումար (դրամ)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    "Income": [100000, 200000, 300000, 400000, 600000],
    "Debt": [1000000, 800000, 600000, 400000, 200000],
    "Savings": [0, 50000, 120000, 250000, 500000]
}

df = pd.DataFrame(data)

print(df)

# Գրաֆիկ
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Income"], marker="o", label="Income")
plt.plot(df["Year"], df["Debt"], marker="x", label="Debt")
plt.plot(df["Year"], df["Savings"], marker="s", label="Savings")

plt.title("Քո ֆինանսական ճանապարհը (2021–2025)")
plt.xlabel("Տարի")
plt.ylabel("Գումար (դրամ)")
plt.legend()
plt.grid(True)
plt.show()

```
